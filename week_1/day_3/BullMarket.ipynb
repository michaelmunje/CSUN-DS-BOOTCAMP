```python
import pandas as pd
df = pd.read_csv("BULLMARKET.csv")
df.describe()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
# df.drop(columns=["Date"], inplace=True)
df[:100]
```

```python
import matplotlib.pyplot as plt
rr = pd.plotting.scatter_matrix(df, figsize=(15,15), 
                                marker="o",s=10,c="black")
fig=plt.gcf()
fig.show()
```

```python
#X = df.drop(columns=["Volume"]).values
X = df["Date"].values.reshape(-1,1)
Y = df['Open'].values
```

```python
import sklearn
from sklearn.model_selection import train_test_split
XTRAIN, XTEST, YTRAIN, YTEST= train_test_split(X,Y, shuffle=False)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.
from sklearn.metrics import mean_squared_error, r2_score

r = LinearRegression().fit(XTRAIN,YTRAIN)
P=r.predict(XTEST)
```

```python
R2=r2_score(YTEST,P)
MSE = mean_squared_error(YTEST,P)
print(R2,MSE)
```

```python
xplot=[[2009],[2020]]
yplot=r.predict(xplot)
plt.scatter(XTRAIN, YTRAIN, s=5, color="blue",label="train")
plt.scatter(XTEST, YTEST,s=5,color="red", label="test")
plt.plot(xplot,yplot, color="orange", label="Regression", linewidth=5)
plt.grid()
plt.legend()
fig=plt.gcf()
fig.set_size_inches(12,12)
plt.show()
```
